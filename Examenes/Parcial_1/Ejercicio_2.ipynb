*total datos:3x365 = 1.095 dias*

    enero21° / 11°
    febrero21° / 11°
    marzo21° / 11°
    abril21° / 9°

    junio21° / 6°
    julio20° / 5°
    agosto22° / 7°
    septiembre22° / 8°
    octubre23° / 10°
    noviembre23° / 11°
    diciembre22° / 11°

```python
import random
import numpy as np
#mes 1
def generarTemperaturas():
    año1=[]
    for i in range(365):
        if(i+1<=90): #enero febrero marzo
            año1.append(random.uniform(11, 21))
        if(i+1<=120 and i+1>90):#abril
            año1.append(random.uniform(9, 21))
        if(i+1<=150 and i+1>120):#mayo        
            año1.append(random.uniform(8, 21))
        if(i+1<=180 and i+1>150):#junio
            año1.append(random.uniform(6, 21))
        if(i+1<=210 and i+1>180):#julio
            año1.append(random.uniform(5, 20))
        if(i+1<=240 and i+1>210):#agosto 
            año1.append(random.uniform(7, 22))
        if(i+1<=270 and i+1>240):#sep
            año1.append(random.uniform(8, 22))
        if(i+1<=300 and i+1>270):#octubre
            año1.append(random.uniform(10, 23))
        if(i+1<=330 and i+1>300):#noviembre
            año1.append(random.uniform(11, 23))
        if(i+1<=365 and i+1>330):#diciembre
            año1.append(random.uniform(11, 22))
    return np.array(año1)
```

```python
año_1=generarTemperaturas()
año_1.shape
año_2=generarTemperaturas()
año_2.shape
año_3=generarTemperaturas()
año_3.shape
```

```python
dataset = np.concatenate((año_1, año_3, año_3)) 
```
